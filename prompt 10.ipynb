```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#setting up variables for rng hypergeometric
rng = np.random.default_rng()
ngood, nbad, nsamp = 50, 5000, 1000
#set up for histogram
plt.hist(rng.hypergeometric(ngood, nbad, nsamp, 1000), bins=100)
#graphing
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Hyper Geometrically Random numbers')
plt.savefig('Hyper Geometrically Random numbers.pdf', format='pdf')
```
